```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style = "whitegrid")
```

```python
tips = sns.load_dataset('tips')
print(tips.head())
```

```python
plt.figure(figsize=(8,4))
plt.plot(tips['total_bill'], marker='o', linestyle='-')
plt.title("Total Bill Over Entries")
plt.xlabel("Entry Index")
plt.ylabel("Total Bill")
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.scatter(tips['total_bill'], tips['tip'], alpha=0.7, color='black')
plt.title("Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.histplot(tips['total_bill'], kde=True, bins=20, color = 'black')
plt.title("Distribution of Total Bill")
plt.xlabel("Total bill")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(4, 4))
sns.boxplot(x="day", y="total_bill", data=tips, color="black")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.title("Total Bill by Day")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the 'tips' dataset
tips = sns.load_dataset('tips')

# Create the lmplot
g = sns.lmplot(x='total_bill', y='tip', data=tips, aspect=1.5,)

# Set the title and labels
g.set_axis_labels("Total Bill", "Tip")
g.fig.suptitle("Relationship Between Total Bill and Tip", y=1.03)

# Adjust the figure size
g.fig.set_size_inches(8, 4)

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the 'mpg' dataset
mpg = sns.load_dataset('mpg')

# Create a FacetGrid with histograms
g = sns.FacetGrid(mpg, col="origin")
g.map(plt.hist, "mpg", bins=20, color="skyblue", edgecolor="black")
g.set_axis_labels("Miles per Gallon (MPG)", "Frequency")
g.fig.suptitle("Distribution of MPG by Origin", y=1.03)
plt.show()
```

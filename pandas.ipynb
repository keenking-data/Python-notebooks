```python
import pandas as pd

data = {
    'Product': ['A', 'B', 'A', 'B', 'C'],
    'Sales': [100, 150, 200, 250, 300],
    'Region': ['North', 'South', 'North', 'South', 'East']
}

df = pd.DataFrame(data)
print(df)


```

```python
grouped = df.groupby('Product')['Sales'].sum()
print(grouped)
```

```python
grouped = df.groupby('Product')['Sales'].sum()
print(grouped)
```

```python
aggregated = df.groupby('Product')['Sales'].agg(['sum','mean','max'])
print(aggregated)
```

```python
grouped_multi = df.groupby(['Product','Region'])['Sales'].agg(['sum','mean','max','min'])
print(grouped_multi)
```
